```python
from PIL import Image
from skimage import io
from skimage.transform import radon
import math
from skimage.transform import rotate
from matplotlib import pyplot as plt
import numpy as np
def bresenham(x1,y1,x2, y2):  
    dx = x2 - x1
    dy = y2 - y1

    is_steep = abs(dy) > abs(dx)
    if is_steep:
        x1, y1 = y1, x1
        x2, y2 = y2, x2

    swapped = False
    if x1 > x2:
        x1, x2 = x2, x1
        y1, y2 = y2, y1
        swapped = True

    dx = x2 - x1
    dy = y2 - y1
 
    error = int(dx / 2.0)
    ystep = 1 if y1 < y2 else -1 
    y = y1
    points = []
    for x in range(x1, x2 + 1):
        coord = (y, x) if is_steep else (x, y)
        points.append(coord)
        error -= abs(dy)
        if error < 0:
            y += ystep
            error += dx
    if swapped:
        points.reverse()
    return points
            
img = io.imread('tomograf-zdjecia/Shepp_logan.jpg', as_gray=True)
#img = Image.open('tomograf-zdjecia/Shepp_logan.jpg').convert('LA')

#p=int(math.sqrt(2*img.size))
#for k in range(1,180,8):
 #   theta=k*math.pi/180
  #  a=-math.cos(theta)/math.sin(theta)
   # for pe in range(1,p):
    #    b=pe/math.sin(theta)
     
theta=1 #kat
n=200 #ilosc detektorow
detectors=[]
krok=1
rozpkat=180*math.pi/180
obrazwyj=[]
r=int((math.sqrt(img.shape[0]**2 + img.shape[1]**2))/2)#promien
r=min(img.shape)//2
for i in np.arange(1,180,krok):
    
    theta=i*math.pi/180
    
    x=r*math.cos(theta)
    y=r*math.sin(theta)
    
    for j in range(n):       
        xd=r*math.cos(theta+math.pi-rozpkat/2+j*rozpkat/(n-1))
        yd=r*math.sin(theta+math.pi-rozpkat/2+j*rozpkat/(n-1))
        detectors.append([xd,yd])
            
    for k in range(n):
        
        linia=bresenham(round(detectors[k][0]),round(detectors[k][1]),round(x),round(y))
        #linia=bresenham(round(detectors[k][0]),round(detectors[k][1]),round(detectors[n-k-1][0])+r,round(detectors[n-k-1][1])+r)
        wartosci=[]
        
        for l in linia:
            
            l2x=round(l[0]+r)
            l2y=round(l[1]+r)
            if(l2y>=img.shape[1] or l2x>=img.shape[0]):                
                wartosci.append(0.0)
            else:
                wartosci.append(img[l2x][l2y])
        obrazwyj.append(np.sum(wartosci))  
    detectors=[]
   
    
obrazwyj2=np.array(obrazwyj)
obrazwyj2 = np.reshape(obrazwyj2, (-1, n))
im =Image.fromarray(obrazwyj2)
imgplot = plt.imshow(obrazwyj2)
#io.imshow(obrazwyj2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

from skimage.transform import radon, rescale

image = io.imread('tomograf-zdjecia/Shepp_logan.jpg', as_gray=True)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4.5))

ax1.set_title("Original")
ax1.imshow(image, cmap=plt.cm.Greys_r)

theta = np.linspace(0., 180., max(image.shape), endpoint=False)
sinogram = radon(image, theta=theta, circle=True)
dx, dy = 0.5 * 180.0 / max(image.shape), 0.5 / sinogram.shape[0]
ax2.set_title("Radon transform\n(Sinogram)")
ax2.set_xlabel("Projection angle (deg)")
ax2.set_ylabel("Projection position (pixels)")
ax2.imshow(sinogram, cmap=plt.cm.Greys_r,
           extent=(-dx, 180.0 + dx, -dy, sinogram.shape[0] + dy),
           aspect='auto')

fig.tight_layout()
plt.show()
```
