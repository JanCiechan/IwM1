```python
from PIL import Image
from skimage import io
from skimage.transform import radon
import math
from skimage.transform import rotate
from matplotlib import pyplot as plt
import numpy as np
from scipy import ndimage
```

```python
def bresenham(x1,y1,x2, y2):  
    dx = x2 - x1
    dy = y2 - y1

    is_steep = abs(dy) > abs(dx)
    if is_steep:
        x1, y1 = y1, x1
        x2, y2 = y2, x2

    swapped = False
    if x1 > x2:
        x1, x2 = x2, x1
        y1, y2 = y2, y1
        swapped = True

    dx = x2 - x1
    dy = y2 - y1
 
    error = int(dx / 2.0)
    ystep = 1 if y1 < y2 else -1 
    y = y1
    points = []
    for x in range(x1, x2 + 1):
        coord = (y, x) if is_steep else (x, y)
        points.append(coord)
        error -= abs(dy)
        if error < 0:
            y += ystep
            error += dx
    if swapped:
        points.reverse()
    return points
```

```python

img = io.imread('tomograf-zdjecia/Shepp_logan.jpg', as_gray=True)
theta=1 #kat
n=200 #ilosc detektorow i emiterow
detectors=[]
emiters=[]
krok=1
rozpkat=180*math.pi/180
obrazwyj=[]
print(img.shape[0],img.shape[1])
r2=int(math.sqrt(img.shape[0]**2 + img.shape[1]**2)/2)#promien
r=min(img.shape)//2
odwrotnalinie=[]
ile=0
for i in np.arange(1,180,krok):
    theta=i*math.pi/180
    ile+=1
    x=r*math.cos(theta)
    y=r*math.sin(theta)
    for j in range(n):       
        xd=r*math.cos((theta+math.pi-rozpkat/2)+j*rozpkat/(n-1))
        yd=r*math.sin((theta+math.pi-rozpkat/2)+j*rozpkat/(n-1))
        
        detectors.append([xd,yd])
    for z in detectors:
        xe=z[0]*(-1)
        ye=z[1]*(-1)
        
        emiters.append([xe,ye])
    detectors=detectors[::-1]       
    for k in range(n):        
        linia=bresenham(round(detectors[k][0]+r),round(detectors[k][1]+r),round(emiters[k][0]+r),round(emiters[k][1]+r))        
        wartosci=[]
        odwrotnalinie.append(linia)
        for l in linia:            
            l2x=round(l[0])
            l2y=round(l[1])
            if(l2y>=img.shape[1] or l2x>=img.shape[0]):                
                wartosci.append(0.0)
            else:
                wartosci.append(img[l2x][l2y])
        obrazwyj.append(np.average(wartosci))  
    detectors.clear()
    emiters.clear()
   
    
obrazwyj2=np.array(obrazwyj)
obrazwyj2 = np.reshape(obrazwyj2, (ile, n))
im =Image.fromarray(obrazwyj2)

for i in range(len(obrazwyj2)):
    obrazwyj2[i]=(obrazwyj2[i]-np.min(obrazwyj2))/(np.max(obrazwyj2)-np.min(obrazwyj2))
```

```python
odwrotnawynik=[]
obrazwynikow=np.zeros((img.shape[0]+1,img.shape[1]+1))
licznik=0
for linia in odwrotnalinie:
    for l in linia:
        obrazwynikow[l[0]][l[1]]+=obrazwyj[licznik]
    licznik+=1
for i in range(len(obrazwynikow)):
    obrazwynikow[i]=(obrazwynikow[i]-np.min(obrazwynikow))/(np.max(obrazwynikow)-np.min(obrazwynikow))
```

```python
fig, (ax1, ax2,ax3) = plt.subplots(1, 3, figsize=(10, 4.5),squeeze=True)
ax1.set_title("Original")
ax1.imshow(img, cmap=plt.cm.Greys_r)
ax2.set_title("Radon transform\n(Sinogram)")
ax2.set_xlabel("Projection angle (deg)")
ax2.set_ylabel("Projection position (pixels)")
ax2.imshow( ndimage.rotate(obrazwyj2, 90),cmap=plt.cm.Greys_r)
ax3.set_title("Reverse transform")
ax3.imshow(obrazwynikow,cmap=plt.cm.Greys_r)
fig.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from skimage.transform import radon, rescale

image = io.imread('tomograf-zdjecia/Shepp_logan.jpg', as_gray=True)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4.5))

ax1.set_title("Original")
ax1.imshow(image, cmap=plt.cm.Greys_r)

theta = np.linspace(0., 180., max(image.shape), endpoint=False)
sinogram = radon(image, theta=theta, circle=True)
dx, dy = 0.5 * 180.0 / max(image.shape), 0.5 / sinogram.shape[0]
ax2.set_title("Radon transform\n(Sinogram)")
ax2.set_xlabel("Projection angle (deg)")
ax2.set_ylabel("Projection position (pixels)")
ax2.imshow(sinogram, cmap=plt.cm.Greys_r,
           extent=(-dx, 180.0 + dx, -dy, sinogram.shape[0] + dy),
           aspect='auto')

fig.tight_layout()
plt.show()
```
